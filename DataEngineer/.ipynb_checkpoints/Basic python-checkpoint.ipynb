# dynamic ways

```python
header = ("date", "item", "price")
```

```python
content = [
    ("2020-02-07", "Milk", "26.5"),
    ("2020-02-08", "Water", "7"),
    ("2020-02-08", "Mask", "20"),
    ("2020-02-08", "Lays", "10"),
    ("2020-02-08", "Rice", "6")
]
```

```python
file = open("product.csv", "w")
file.write(",".join(header)+"\n")
for row in content:
    file.write(",".join(row)+"\n")
file.close()
```

```python
with open("product.csv", "r") as f:
    product = f.readlines()
print(product)
```

```python
import pymysql.cursors
import pandas as pd
```

```python
class Config:
    MYSQL_HOST = 'db4free.net'
    MYSQL_PORT = 3306  # default port ของ MySQL คือ 3306
    MYSQL_USER = 'datath'
    MYSQL_PASSWORD = 'DataScience-chillchill'
    MYSQL_DB = 'detraining'
    MYSQL_CHARSET = 'utf8mb4'
```

```python
# Connect to the database
connection = pymysql.connect(host=Config.MYSQL_HOST,
                             port=Config.MYSQL_PORT,
                             user=Config.MYSQL_USER,
                             password=Config.MYSQL_PASSWORD,
                             db=Config.MYSQL_DB,
                             cursorclass=pymysql.cursors.DictCursor)
```

```python
# list all tables
cursor = connection.cursor()
cursor.execute("show tables;")
tables = cursor.fetchall()
cursor.close()
print(tables)
```

```python
with connection.cursor() as cursor:
    sql = "SELECT * FROM online_retail limit 500;"
    cursor.execute(sql)
    result = cursor.fetchall()
```
